**Vertical bar for 5 selected years**

```python
import numpy as np
import matplotlib.pyplot as jig

years = ['2003', '2005', '2007', '2009', '2011']
sachin_runs = [1141, 1234, 1005, 915, 1356]
kohli_runs =  [0,    0,    325, 789, 1380]
sehwag_runs = [1250, 990, 1103, 1025, 987]

x = np.arange(len(years))
width = 0.25

jig.figure(figsize=(8, 5))
jig.bar(x - width, sachin_runs, width, label='Sachin')
jig.bar(x, sehwag_runs, width, label='Sehwag')
jig.bar(x + width, kohli_runs, width, label='Kohli')

jig.xlabel('Year')
jig.ylabel('Runs')
jig.title('Runs over Selected Years')
jig.xticks(x, years)
jig.legend()
jig.grid(axis='y', linestyle='--', alpha=0.5)
jig.tight_layout()
jig.Figure (figsize = (25,21))
jig.show()
```

 **Horizontal Bar for total runs scored in their debut 5 years.**

```python
sachin_5yr = [215, 900, 1100, 1300, 1200]  # 19891993
sehwag_5yr = [250, 850, 1000, 950, 980]    # 19992003
kohli_5yr =  [0, 400, 1020, 1250, 1100]    # 20082012

total_runs = [
    sum(sachin_5yr),
    sum(sehwag_5yr),
    sum(kohli_5yr)
]

players = ['Sachin', 'Sehwag', 'Kohli']
colors = ['blue', 'green', 'Red']

jig.figure(figsize=(7, 4))
jig.barh(players, total_runs, color=colors)
jig.xlabel('Total Runs')
jig.title('Total Runs in First 5 Years of Career')
jig.grid(axis='x', linestyle='--', alpha=0.5)
jig.tight_layout()
jig.show()
```

# Q.no.2.

**Pie Chart for Marketing Shares of OS.**

```python
import matplotlib.pyplot as pie

labels = ['Android', 'iOS', 'Others']
shares = [65, 25, 10]
colors = ['#79DD3F', '#e4ecf1', '#96CAEB'] 

pie.figure(figsize=(5, 5))
pie.pie(
    shares,
    labels=labels,
    colors=colors,
    startangle=90,
    autopct='%1.1f%%'  # Shows percentages
)
pie.title('Mobile OS Market Share (Pie Chart)')
pie.axis('equal') 
pie.show()
```

**Horizontal Bar for Marketing Shares of OS.**

```python
pie.figure(figsize=(6, 3.5))
bars = pie.barh(labels, shares, color=colors)

for bar in bars:
    width = bar.get_width()
    pie.text(width + 1, bar.get_y() + bar.get_height() / 2,
             f'{width}%', va='center', fontsize=10)

pie.title('Mobile OS Market Share (Horizontal Bar Chart)')
pie.xlabel('Market Share (%)')
pie.grid(axis='x', linestyle='--', alpha=0.5)
pie.tight_layout()
pie.show()
```

# Which one is easier to understand ?

**Conclusion:**

*For 23 categories, the pie chart is visually appealing and fine.

*For precise comparison, especially with small values (like "Others"),
 the horizontal bar chart is clearly more effective.
